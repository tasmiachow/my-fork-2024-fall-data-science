### Clean the Data

Perhaps we‚Äôd like to know more another day, but the team is really interested in just the relationship between calories (`calories_day`) and weight. ‚Ä¶and maybe gender.

Can you remove the other columns?

```python
# Remove ‚Äòem.
df_filter= df_food[['calories_day', 'weight', 'Gender']]
df_filter.head()
```

What about `NaN`s? How many are there?

```python
# Count ‚Äòem.

df_filter.isnull().sum()
```

We gotta remove those `NaN`s‚Äïthe entire row.

```python
# Drop ‚Äòem.

df_filter = df_filter.dropna()

df_filter.head()
```

But what about the weird non-numeric values in the column obviously meant for numeric data?

Notice the data type of that column from when you got the types of all the columns?

If only we could convert the column to a numeric type and drop the rows with invalid values. ü§î

```python
# Fix that.

print(df_filter.dtypes)
df_filter['calories_day'] = pd.to_numeric(df_filter['calories_day'], errors='coerce')
df_filter['weight'] = pd.to_numeric(df_filter['weight'], errors='coerce')
df_filter= df_filter.dropna(subset=['calories_day', 'weight'])
df_filter.head()
```

Now this data seems reasonably clean for our purposes! üòÅ

```python
# Savey save!
output_path = 'data/cleaned_food_data.csv'
df_filter.to_csv(output_path, index=False)
```

### Clean the Data

Well, we can‚Äôt get our answers with what we currently have, so you‚Äôll have to make some changes.

Let‚Äôs not worry about anything below the first 5 values for now. Convert the top 5 to a single canonical value‚Äïsay, `'US'`, which is nice and short.

```python
# Replace them all with 'US'.
print(df_salary_usd['country'].value_counts().head(5).index)
df_salary_usd = df_salary_usd.copy()
# Print top 5 countries to verify
print(df_salary_usd['country'].value_counts().head(5).index)

# Get top 5 countries
top_5_countries = df_salary_usd['country'].value_counts().head(5).index

# Clean the 'country' column if necessary (e.g., strip whitespace)
df_salary_usd['country'] = df_salary_usd['country'].str.strip()

# Replace these top 5 values with 'US'
df_salary_usd.loc[df_salary_usd['country'].isin(top_5_countries), 'country'] = 'US'

# Verify the changes
print(df_salary_usd['country'].unique())
```

It‚Äôs looking good so far. Let‚Äôs find out the minimum, mean, and maximum (in that order) salary by state, sorted by the mean in descending order.

```python
# Find the minimum, mean, and maximum salary in USD by U.S. state.
salary_stats = df_salary_usd.groupby('state')['salary'].agg(['min', 'mean', 'max'])

# Sort by the mean salary in descending order
salary_stats_sorted = salary_stats.sort_values(by='mean', ascending=False)

# Display the result
print(salary_stats_sorted)
```

Well, pooh! We forgot that `salary` isn‚Äôt numeric. Something wrong must be fixed.

```python
# Fix it.
df_salary_usd = df_salary_usd.copy()

# Convert the 'salary' column to numeric using .loc
df_salary_usd.loc[:, 'salary'] = pd.to_numeric(df_salary_usd['salary'], errors='coerce')

# Optionally, drop rows with NaN salaries (if applicable)
df_salary_usd = df_salary_usd.dropna(subset=['salary'])

df_salary_usd.head()
```

Let‚Äôs try that again.

```python
# Try it again. Yeah!

salary_stats = df_salary_usd.groupby('state')['salary'].agg(['min', 'mean', 'max'])

# Sort by the mean salary in descending order
salary_stats_sorted = salary_stats.sort_values(by='mean', ascending=False)

# Display the result
print(salary_stats_sorted)
```

That did the trick! Now let‚Äôs narrow this to data 2021 and 2022 just because (lel). *(Hint: that timestamp column may not be a temporal type right now.)*

```python
# Filter the data to within 2021, 2022, or 2023, saving the DataFrame to a new variable, and generate the summary again.

# Ensure 'timestamp' is a datetime type
df_salary_usd['timestamp'] = pd.to_datetime(df_salary_usd['timestamp'], errors='coerce')

# Filter rows for the years 2021, 2022, or 2023
filtered_data = df_salary_usd[df_salary_usd['timestamp'].dt.year.isin([2021, 2022, 2023])]

# Group by state and calculate min, mean, and max salary
salary_stats_filtered = filtered_data.groupby('state')['salary'].agg(['min', 'mean', 'max'])

# Sort by the mean salary in descending order
salary_stats_filtered_sorted = salary_stats_filtered.sort_values(by='mean', ascending=False)

# Display the result
print(salary_stats_filtered_sorted)
```
